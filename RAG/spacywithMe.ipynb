```python
import spacy
import pandas as pd      
import numpy as np         
import seaborn as sns          
import matplotlib.pyplot as plt 
```

```python
# we downloaded small model 
# - en: english
# - sm: size of model
# - web: type of text pipeline ws trained on 
# - core: capabilities

nlp = spacy.load("en_core_web_sm")
```

```python
data_text = "mosquitoes.txt"

with open(data_text, "r", encoding="utf-8") as file:
    content = file.read()
    
print(content)
```

```python
import re
pattern = re.split(r'(?<=[.!?])\s+', content)

for s in pattern:
    print(s)
```

```python
import re
pattern = re.split(r'\n\s*\n', content)

for s in pattern:
    print(s)
```

```python
doc = nlp(content)

print(doc)
```

```python
for token in content.split()[:10]:
    print(token)
```

```python
for token in doc[:10]:
    print(token)
```

```python
# sentence boundary detection

for sent in doc.sents:
    print(sent)
```

```python
sentence1 = list(doc.sents)[0]
print(sentence1)
```
