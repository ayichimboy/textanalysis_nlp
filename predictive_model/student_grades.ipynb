```python
# Import Libraries

import numpy as np                    
import pandas as pd                
import matplotlib.pyplot as plt                   
import seaborn as sns                          
```

```python
# Load the data 
data = pd.read_csv("StudentPerformance.csv")

data.head()
```

```python
data.info()
```

```python
data.columns = data.columns.str.replace(" ", "_").str.lower().str.strip()
data.columns
```

```python
for col in data.columns:
    print(col)
```

```python
for unique in data['extracurricular_activities'].unique():
    print(unique)
```

```python
sns.pairplot(data, hue="extracurricular_activities")
plt.show()
```

```python
df_num = data.select_dtypes(include=["float64", "int64"])

df_corr = pd.DataFrame(df_num.corr())
df_corr
```

```python
sns.heatmap(df_corr, fmt=".2f", annot=True)
plt.show()
```
